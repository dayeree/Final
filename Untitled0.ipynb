```python
import os
import json

# =========================================================
# ✅ 이미지와 json 매칭
# =========================================================

# 예: 이미지 파일 경로에서 파일명만 추출
sorted_image_files = sorted([
    os.path.basename(f) for f in os.listdir('/content/bdd100k_images/val')
    if f.lower().endswith(('.jpg', '.jpeg', '.png'))
])

# 예: JSON 디렉토리에서 파일 열고 이미지 이름 추출
json_dir = '/content/bdd100k_labels/bdd100k/labels/100k/val'
json_files = []

for fname in os.listdir(json_dir):
    if fname.endswith('.json'):
        with open(os.path.join(json_dir, fname), 'r') as f:
            data = json.load(f)
            if 'name' in data:
                json_files.append(data['name'] + '.jpg')  # ✅ 확장자 맞춰주기

# ✅ 집합 변환
image_set = set(sorted_image_files)
json_set = set(json_files)

# 1. 매칭된 파일
matched = image_set & json_set

# 2. 이미지만 있고 라벨 없는 경우
image_only = image_set - json_set

# 3. 라벨만 있고 이미지 없는 경우
json_only = json_set - image_set

# ✅ 결과 출력
print(f"✅ JSON과 매칭되는 이미지 수: {len(matched)}개")
print(f"❌ JSON 없는 이미지 수: {len(image_only)}개")
print(f"❌ 이미지 없는 JSON 수: {len(json_only)}개")

# 🔍 예시 출력
print("\n📌 JSON 없는 이미지 예시:")
print(list(image_only)[:5])

print("\n📌 이미지 없는 JSON 예시:")
print(list(json_only)[:5])
```

```python
# matched = image_set & json_set
matched_files = sorted(list(image_set & json_set))  # 리스트로 변환 + 정렬

# 비율 설정
total = len(matched_files)
train_ratio, val_ratio, test_ratio = 0.7, 0.1, 0.2

train_end = int(total * train_ratio)
val_end = train_end + int(total * val_ratio)

train_set = matched_files[:train_end]
val_set = matched_files[train_end:val_end]
test_set = matched_files[val_end:]

print(f"📦 분할 결과:")
print(f"  - Train: {len(train_set)}개")
print(f"  - Val:   {len(val_set)}개")
print(f"  - Test:  {len(test_set)}개")
```

```python
import os
import json
import cv2
from tqdm import tqdm

# ✅ 경로 설정 (train 기준)
image_dir = '/content/processed_data/train/images'
json_dir = '/content/processed_data/train/labels'
output_dir = '/content/drive/MyDrive/cropped_signs_train'  # ⬅️ train 결과 저장 디렉토리
os.makedirs(output_dir, exist_ok=True)

# JSON 파일 리스트 정렬
json_files = sorted([f for f in os.listdir(json_dir) if f.endswith('.json')])

# 크롭 이미지 저장을 위한 카운트
count = 0

# 각 JSON 파일에 대해 처리
for jf in tqdm(json_files, desc='✂️ train traffic sign crop 중'):
    json_path = os.path.join(json_dir, jf)
    image_name = os.path.splitext(jf)[0] + '.jpg'
    image_path = os.path.join(image_dir, image_name)

    if not os.path.exists(image_path):
        continue

    image = cv2.imread(image_path)
    if image is None:
        continue

    with open(json_path, 'r') as f:
        data = json.load(f)

    for frame in data.get('frames', []):
        for obj in frame.get('objects', []):
            if obj.get('category') != 'traffic sign':
                continue

            attr = obj.get('attributes', {})
            if attr.get('occluded', False) or attr.get('truncated', False):
                continue

            box = obj.get('box2d', {})
            x1, y1 = int(box.get('x1', 0)), int(box.get('y1', 0))
            x2, y2 = int(box.get('x2', 0)), int(box.get('y2', 0))
            w, h = x2 - x1, y2 - y1

            if w < 15 or h < 15:
                continue

            cropped = image[y1:y2, x1:x2]
            if cropped.size == 0:
                continue

            save_path = os.path.join(output_dir, f'{os.path.splitext(jf)[0]}_{count}.jpg')
            cv2.imwrite(save_path, cropped)
            count += 1

# ✅ 최종 결과 출력
print(f"\n✅ 총 {count}개의 traffic sign crop 이미지가 train 세트에서 저장되었습니다.")
```

```python
import os
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm

# 경로 설정
crop_dir = '/content/drive/MyDrive/cropped_signs_train'

# 필터링 기준
MIN_SIZE = 50
MAX_ASPECT_RATIO = 3.0

filtered_images = []

print("🧹 불필요한 이미지 필터링 중...")

for fname in tqdm(os.listdir(crop_dir)):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    path = os.path.join(crop_dir, fname)
    try:
        img = Image.open(path)
        w, h = img.size
        aspect_ratio = max(w / h, h / w)

        if w < MIN_SIZE or h < MIN_SIZE or aspect_ratio > MAX_ASPECT_RATIO:
            os.remove(path)
        else:
            filtered_images.append(fname)

    except UnidentifiedImageError:
        os.remove(path)

print(f"\n✅ 최종 남은 이미지 수: {len(filtered_images)}개")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil

# =========================================================
# ✅ 유사도 0.5 기준으로 train set에 대해서 분류
# =========================================================

# ====== 경로 설정 ======
input_dir = '/content/drive/MyDrive/cropped_signs_train'  # ✅ 경로 확인
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/train_signs'
csv_output_path = '/content/drive/MyDrive/train1.csv'

# ====== 디렉토리 존재 확인 ======
assert os.path.exists(input_dir), f"❌ 입력 디렉토리 없음: {input_dir}"
assert os.path.exists(reference_dir), f"❌ 기준 디렉토리 없음: {reference_dir}"

# ====== 폴더 준비 ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)

# ====== 모델 준비 ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)  # ✅ 최신 권장 방식 사용
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== 클래스 임베딩 생성 ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== 이미지 분류 ======
results = []

for file in tqdm(os.listdir(input_dir), desc="🧠 유사도 기반 분류 중"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    # cosine similarity 계산
    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    # threshold에 따라 unknown 처리
    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== 결과 저장 ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\n✅ CSV 저장 완료: {csv_output_path}")
```

```python
import os
import json
import cv2
from tqdm import tqdm

# ✅ 경로 설정 (val 기준)
image_dir = '/content/processed_data/val/images'
json_dir = '/content/processed_data/val/labels'
output_dir = '/content/drive/MyDrive/cropped_signs_val'  # ⬅️ train 결과 저장 디렉토리
os.makedirs(output_dir, exist_ok=True)

# JSON 파일 리스트 정렬
json_files = sorted([f for f in os.listdir(json_dir) if f.endswith('.json')])

# 크롭 이미지 저장을 위한 카운트
count = 0

# 각 JSON 파일에 대해 처리
for jf in tqdm(json_files, desc='✂️ val traffic sign crop 중'):
    json_path = os.path.join(json_dir, jf)
    image_name = os.path.splitext(jf)[0] + '.jpg'
    image_path = os.path.join(image_dir, image_name)

    if not os.path.exists(image_path):
        continue

    image = cv2.imread(image_path)
    if image is None:
        continue

    with open(json_path, 'r') as f:
        data = json.load(f)

    for frame in data.get('frames', []):
        for obj in frame.get('objects', []):
            if obj.get('category') != 'traffic sign':
                continue

            attr = obj.get('attributes', {})
            if attr.get('occluded', False) or attr.get('truncated', False):
                continue

            box = obj.get('box2d', {})
            x1, y1 = int(box.get('x1', 0)), int(box.get('y1', 0))
            x2, y2 = int(box.get('x2', 0)), int(box.get('y2', 0))
            w, h = x2 - x1, y2 - y1

            if w < 15 or h < 15:
                continue

            cropped = image[y1:y2, x1:x2]
            if cropped.size == 0:
                continue

            save_path = os.path.join(output_dir, f'{os.path.splitext(jf)[0]}_{count}.jpg')
            cv2.imwrite(save_path, cropped)
            count += 1

# ✅ 최종 결과 출력
print(f"\n✅ 총 {count}개의 traffic sign crop 이미지가 val 세트에서 저장되었습니다.")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil


# ====== 경로 설정 (val 기준) ======
input_dir = '/content/drive/MyDrive/cropped_signs_val'  # ✅ val 경로
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/val_signs'  # ✅ val 결과 저장 디렉토리
csv_output_path = '/content/drive/MyDrive/val1.csv'  # ✅ val CSV 이름

# ====== 디렉토리 존재 확인 ======
assert os.path.exists(input_dir), f"❌ 입력 디렉토리 없음: {input_dir}"
assert os.path.exists(reference_dir), f"❌ 기준 디렉토리 없음: {reference_dir}"

# ====== 폴더 준비 ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)


# ====== 모델 준비 ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)  # ✅ 최신 권장 방식 사용
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== 클래스 임베딩 생성 ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== 이미지 분류 ======
results = []

for file in tqdm(os.listdir(input_dir), desc="🧠 유사도 기반 분류 중"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    # cosine similarity 계산
    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    # threshold에 따라 unknown 처리
    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== 결과 저장 ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\n✅ CSV 저장 완료: {csv_output_path}")
```

```python
import os
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm

# 경로 설정
crop_dir = '/content/drive/MyDrive/cropped_signs_test'

# 필터링 기준
MIN_SIZE = 50
MAX_ASPECT_RATIO = 3.0

filtered_images = []

print("🧹 불필요한 이미지 필터링 중...")

for fname in tqdm(os.listdir(crop_dir)):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    path = os.path.join(crop_dir, fname)
    try:
        img = Image.open(path)
        w, h = img.size
        aspect_ratio = max(w / h, h / w)

        if w < MIN_SIZE or h < MIN_SIZE or aspect_ratio > MAX_ASPECT_RATIO:
            os.remove(path)
        else:
            filtered_images.append(fname)

    except UnidentifiedImageError:
        os.remove(path)

print(f"\n✅ 최종 남은 이미지 수: {len(filtered_images)}개")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil

# =========================================================
# ✅ 유사도 0.5 기준으로 test set에 대해서 분류
# =========================================================

# ====== 경로 설정 ======
input_dir = '/content/drive/MyDrive/cropped_signs_test'  # ✅ test 기준 경로
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/test_signs'  # ✅ 결과 저장 디렉토리
csv_output_path = '/content/drive/MyDrive/test1.csv'  # ✅ CSV 파일명

# ====== 디렉토리 존재 확인 ======
assert os.path.exists(input_dir), f"❌ 입력 디렉토리 없음: {input_dir}"
assert os.path.exists(reference_dir), f"❌ 기준 디렉토리 없음: {reference_dir}"

# ====== 폴더 준비 ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)

# ====== 모델 준비 ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== 클래스 임베딩 생성 ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== 이미지 분류 ======
results = []

for file in tqdm(os.listdir(input_dir), desc="🧠 test 세트 유사도 분류 중"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== 결과 저장 ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\n✅ CSV 저장 완료: {csv_output_path}")
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
import seaborn as sns
import matplotlib.pyplot as plt

# 경로 설정 (✅ test 기준으로 변경)
manual_label_path = '/content/drive/MyDrive/test1_final_match.xlsx'  # 수동 라벨링 파일
prediction_path = '/content/drive/MyDrive/test1.csv'  # 모델 예측 결과

# 🔹 엑셀 (수동 정답 라벨) 불러오기
df_manual = pd.read_excel(manual_label_path)[['filename', 'match']]
df_manual.rename(columns={'match': 'true_label'}, inplace=True)

# 🔹 CSV (모델 예측 결과) 불러오기
df_pred = pd.read_csv(prediction_path)[['filename', 'class']]
df_pred.rename(columns={'class': 'predicted_label'}, inplace=True)

# 🔹 병합
df = pd.merge(df_manual, df_pred, on='filename', how='inner')

# 🔹 혼동 행렬 계산
labels = sorted(df['true_label'].unique())
cm = confusion_matrix(df['true_label'], df['predicted_label'], labels=labels)

# 🔹 혼동 행렬 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(cm, index=labels, columns=labels), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix (Test Set)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# 🔹 정밀도, 재현율, F1 계산
precision, recall, f1, _ = precision_recall_fscore_support(df['true_label'], df['predicted_label'], labels=labels, average=None)

print("📊 클래스별 Precision, Recall, F1 Score (Test Set):")
for label, p, r, f in zip(labels, precision, recall, f1):
    print(f"{label:<10} | Precision: {p:.2f} | Recall: {r:.2f} | F1 Score: {f:.2f}")

# 🔹 전체 평균
precision_macro = precision.mean()
recall_macro = recall.mean()
f1_macro = f1.mean()

print(f"\n📈 전체 평균 Precision: {precision_macro:.2f}")
print(f"📈 전체 평균 Recall: {recall_macro:.2f}")
print(f"📈 전체 평균 F1 Score: {f1_macro:.2f}")
```
