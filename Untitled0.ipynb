```python
import os
import json

# =========================================================
# âœ… ì´ë¯¸ì§€ì™€ json ë§¤ì¹­
# =========================================================

# ì˜ˆ: ì´ë¯¸ì§€ íŒŒì¼ ê²½ë¡œì—ì„œ íŒŒì¼ëª…ë§Œ ì¶”ì¶œ
sorted_image_files = sorted([
    os.path.basename(f) for f in os.listdir('/content/bdd100k_images/val')
    if f.lower().endswith(('.jpg', '.jpeg', '.png'))
])

# ì˜ˆ: JSON ë””ë ‰í† ë¦¬ì—ì„œ íŒŒì¼ ì—´ê³  ì´ë¯¸ì§€ ì´ë¦„ ì¶”ì¶œ
json_dir = '/content/bdd100k_labels/bdd100k/labels/100k/val'
json_files = []

for fname in os.listdir(json_dir):
    if fname.endswith('.json'):
        with open(os.path.join(json_dir, fname), 'r') as f:
            data = json.load(f)
            if 'name' in data:
                json_files.append(data['name'] + '.jpg')  # âœ… í™•ì¥ì ë§ì¶°ì£¼ê¸°

# âœ… ì§‘í•© ë³€í™˜
image_set = set(sorted_image_files)
json_set = set(json_files)

# 1. ë§¤ì¹­ëœ íŒŒì¼
matched = image_set & json_set

# 2. ì´ë¯¸ì§€ë§Œ ìˆê³  ë¼ë²¨ ì—†ëŠ” ê²½ìš°
image_only = image_set - json_set

# 3. ë¼ë²¨ë§Œ ìˆê³  ì´ë¯¸ì§€ ì—†ëŠ” ê²½ìš°
json_only = json_set - image_set

# âœ… ê²°ê³¼ ì¶œë ¥
print(f"âœ… JSONê³¼ ë§¤ì¹­ë˜ëŠ” ì´ë¯¸ì§€ ìˆ˜: {len(matched)}ê°œ")
print(f"âŒ JSON ì—†ëŠ” ì´ë¯¸ì§€ ìˆ˜: {len(image_only)}ê°œ")
print(f"âŒ ì´ë¯¸ì§€ ì—†ëŠ” JSON ìˆ˜: {len(json_only)}ê°œ")

# ğŸ” ì˜ˆì‹œ ì¶œë ¥
print("\nğŸ“Œ JSON ì—†ëŠ” ì´ë¯¸ì§€ ì˜ˆì‹œ:")
print(list(image_only)[:5])

print("\nğŸ“Œ ì´ë¯¸ì§€ ì—†ëŠ” JSON ì˜ˆì‹œ:")
print(list(json_only)[:5])
```

```python
import os
import json
import cv2
from tqdm import tqdm

# âœ… ê²½ë¡œ ì„¤ì • (train ê¸°ì¤€)
image_dir = '/content/processed_data/train/images'
json_dir = '/content/processed_data/train/labels'
output_dir = '/content/drive/MyDrive/cropped_signs_train'  # â¬…ï¸ train ê²°ê³¼ ì €ì¥ ë””ë ‰í† ë¦¬
os.makedirs(output_dir, exist_ok=True)

# JSON íŒŒì¼ ë¦¬ìŠ¤íŠ¸ ì •ë ¬
json_files = sorted([f for f in os.listdir(json_dir) if f.endswith('.json')])

# í¬ë¡­ ì´ë¯¸ì§€ ì €ì¥ì„ ìœ„í•œ ì¹´ìš´íŠ¸
count = 0

# ê° JSON íŒŒì¼ì— ëŒ€í•´ ì²˜ë¦¬
for jf in tqdm(json_files, desc='âœ‚ï¸ train traffic sign crop ì¤‘'):
    json_path = os.path.join(json_dir, jf)
    image_name = os.path.splitext(jf)[0] + '.jpg'
    image_path = os.path.join(image_dir, image_name)

    if not os.path.exists(image_path):
        continue

    image = cv2.imread(image_path)
    if image is None:
        continue

    with open(json_path, 'r') as f:
        data = json.load(f)

    for frame in data.get('frames', []):
        for obj in frame.get('objects', []):
            if obj.get('category') != 'traffic sign':
                continue

            attr = obj.get('attributes', {})
            if attr.get('occluded', False) or attr.get('truncated', False):
                continue

            box = obj.get('box2d', {})
            x1, y1 = int(box.get('x1', 0)), int(box.get('y1', 0))
            x2, y2 = int(box.get('x2', 0)), int(box.get('y2', 0))
            w, h = x2 - x1, y2 - y1

            if w < 15 or h < 15:
                continue

            cropped = image[y1:y2, x1:x2]
            if cropped.size == 0:
                continue

            save_path = os.path.join(output_dir, f'{os.path.splitext(jf)[0]}_{count}.jpg')
            cv2.imwrite(save_path, cropped)
            count += 1

# âœ… ìµœì¢… ê²°ê³¼ ì¶œë ¥
print(f"\nâœ… ì´ {count}ê°œì˜ traffic sign crop ì´ë¯¸ì§€ê°€ train ì„¸íŠ¸ì—ì„œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import os
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm

# ê²½ë¡œ ì„¤ì •
crop_dir = '/content/drive/MyDrive/cropped_signs_train'

# í•„í„°ë§ ê¸°ì¤€
MIN_SIZE = 50
MAX_ASPECT_RATIO = 3.0

filtered_images = []

print("ğŸ§¹ ë¶ˆí•„ìš”í•œ ì´ë¯¸ì§€ í•„í„°ë§ ì¤‘...")

for fname in tqdm(os.listdir(crop_dir)):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    path = os.path.join(crop_dir, fname)
    try:
        img = Image.open(path)
        w, h = img.size
        aspect_ratio = max(w / h, h / w)

        if w < MIN_SIZE or h < MIN_SIZE or aspect_ratio > MAX_ASPECT_RATIO:
            os.remove(path)
        else:
            filtered_images.append(fname)

    except UnidentifiedImageError:
        os.remove(path)

print(f"\nâœ… ìµœì¢… ë‚¨ì€ ì´ë¯¸ì§€ ìˆ˜: {len(filtered_images)}ê°œ")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil

# =========================================================
# âœ… ìœ ì‚¬ë„ 0.5 ê¸°ì¤€ìœ¼ë¡œ train setì— ëŒ€í•´ì„œ ë¶„ë¥˜
# =========================================================

# ====== ê²½ë¡œ ì„¤ì • ======
input_dir = '/content/drive/MyDrive/cropped_signs_train'  # âœ… ê²½ë¡œ í™•ì¸
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/train_signs'
csv_output_path = '/content/drive/MyDrive/train1.csv'

# ====== ë””ë ‰í† ë¦¬ ì¡´ì¬ í™•ì¸ ======
assert os.path.exists(input_dir), f"âŒ ì…ë ¥ ë””ë ‰í† ë¦¬ ì—†ìŒ: {input_dir}"
assert os.path.exists(reference_dir), f"âŒ ê¸°ì¤€ ë””ë ‰í† ë¦¬ ì—†ìŒ: {reference_dir}"

# ====== í´ë” ì¤€ë¹„ ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)

# ====== ëª¨ë¸ ì¤€ë¹„ ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)  # âœ… ìµœì‹  ê¶Œì¥ ë°©ì‹ ì‚¬ìš©
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== í´ë˜ìŠ¤ ì„ë² ë”© ìƒì„± ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== ì´ë¯¸ì§€ ë¶„ë¥˜ ======
results = []

for file in tqdm(os.listdir(input_dir), desc="ğŸ§  ìœ ì‚¬ë„ ê¸°ë°˜ ë¶„ë¥˜ ì¤‘"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    # cosine similarity ê³„ì‚°
    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    # thresholdì— ë”°ë¼ unknown ì²˜ë¦¬
    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== ê²°ê³¼ ì €ì¥ ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\nâœ… CSV ì €ì¥ ì™„ë£Œ: {csv_output_path}")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil


# ====== ê²½ë¡œ ì„¤ì • (val ê¸°ì¤€) ======
input_dir = '/content/drive/MyDrive/cropped_signs_val'  # âœ… val ê²½ë¡œ
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/val_signs'  # âœ… val ê²°ê³¼ ì €ì¥ ë””ë ‰í† ë¦¬
csv_output_path = '/content/drive/MyDrive/val1.csv'  # âœ… val CSV ì´ë¦„

# ====== ë””ë ‰í† ë¦¬ ì¡´ì¬ í™•ì¸ ======
assert os.path.exists(input_dir), f"âŒ ì…ë ¥ ë””ë ‰í† ë¦¬ ì—†ìŒ: {input_dir}"
assert os.path.exists(reference_dir), f"âŒ ê¸°ì¤€ ë””ë ‰í† ë¦¬ ì—†ìŒ: {reference_dir}"

# ====== í´ë” ì¤€ë¹„ ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)


# ====== ëª¨ë¸ ì¤€ë¹„ ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)  # âœ… ìµœì‹  ê¶Œì¥ ë°©ì‹ ì‚¬ìš©
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== í´ë˜ìŠ¤ ì„ë² ë”© ìƒì„± ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== ì´ë¯¸ì§€ ë¶„ë¥˜ ======
results = []

for file in tqdm(os.listdir(input_dir), desc="ğŸ§  ìœ ì‚¬ë„ ê¸°ë°˜ ë¶„ë¥˜ ì¤‘"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    # cosine similarity ê³„ì‚°
    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    # thresholdì— ë”°ë¼ unknown ì²˜ë¦¬
    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== ê²°ê³¼ ì €ì¥ ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\nâœ… CSV ì €ì¥ ì™„ë£Œ: {csv_output_path}")
```

```python
import os
import json
import cv2
from tqdm import tqdm

# âœ… ê²½ë¡œ ì„¤ì • (test ê¸°ì¤€)
image_dir = '/content/processed_data/test/images'
json_dir = '/content/processed_data/test/labels'
output_dir = '/content/drive/MyDrive/cropped_signs_test'  # â¬…ï¸ test ê²°ê³¼ ì €ì¥ ë””ë ‰í† ë¦¬
os.makedirs(output_dir, exist_ok=True)

# JSON íŒŒì¼ ë¦¬ìŠ¤íŠ¸ ì •ë ¬
json_files = sorted([f for f in os.listdir(json_dir) if f.endswith('.json')])

# í¬ë¡­ ì´ë¯¸ì§€ ì €ì¥ì„ ìœ„í•œ ì¹´ìš´íŠ¸
count = 0

# ê° JSON íŒŒì¼ì— ëŒ€í•´ ì²˜ë¦¬
for jf in tqdm(json_files, desc='âœ‚ï¸ test traffic sign crop ì¤‘'):
    json_path = os.path.join(json_dir, jf)
    image_name = os.path.splitext(jf)[0] + '.jpg'
    image_path = os.path.join(image_dir, image_name)

    if not os.path.exists(image_path):
        continue

    image = cv2.imread(image_path)
    if image is None:
        continue

    with open(json_path, 'r') as f:
        data = json.load(f)

    for frame in data.get('frames', []):
        for obj in frame.get('objects', []):
            if obj.get('category') != 'traffic sign':
                continue

            attr = obj.get('attributes', {})
            if attr.get('occluded', False) or attr.get('truncated', False):
                continue

            box = obj.get('box2d', {})
            x1, y1 = int(box.get('x1', 0)), int(box.get('y1', 0))
            x2, y2 = int(box.get('x2', 0)), int(box.get('y2', 0))
            w, h = x2 - x1, y2 - y1

            if w < 15 or h < 15:
                continue

            cropped = image[y1:y2, x1:x2]
            if cropped.size == 0:
                continue

            save_path = os.path.join(output_dir, f'{os.path.splitext(jf)[0]}_{count}.jpg')
            cv2.imwrite(save_path, cropped)
            count += 1

# âœ… ìµœì¢… ê²°ê³¼ ì¶œë ¥
print(f"\nâœ… ì´ {count}ê°œì˜ traffic sign crop ì´ë¯¸ì§€ê°€ test ì„¸íŠ¸ì—ì„œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

```python
import os
from PIL import Image, UnidentifiedImageError
from tqdm import tqdm

# ê²½ë¡œ ì„¤ì •
crop_dir = '/content/drive/MyDrive/cropped_signs_test'

# í•„í„°ë§ ê¸°ì¤€
MIN_SIZE = 50
MAX_ASPECT_RATIO = 3.0

filtered_images = []

print("ğŸ§¹ ë¶ˆí•„ìš”í•œ ì´ë¯¸ì§€ í•„í„°ë§ ì¤‘...")

for fname in tqdm(os.listdir(crop_dir)):
    if not fname.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    path = os.path.join(crop_dir, fname)
    try:
        img = Image.open(path)
        w, h = img.size
        aspect_ratio = max(w / h, h / w)

        if w < MIN_SIZE or h < MIN_SIZE or aspect_ratio > MAX_ASPECT_RATIO:
            os.remove(path)
        else:
            filtered_images.append(fname)

    except UnidentifiedImageError:
        os.remove(path)

print(f"\nâœ… ìµœì¢… ë‚¨ì€ ì´ë¯¸ì§€ ìˆ˜: {len(filtered_images)}ê°œ")
```

```python
import os
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet18, ResNet18_Weights
from PIL import Image, UnidentifiedImageError
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import shutil

# =========================================================
# âœ… ìœ ì‚¬ë„ 0.5 ê¸°ì¤€ìœ¼ë¡œ test setì— ëŒ€í•´ì„œ ë¶„ë¥˜
# =========================================================

# ====== ê²½ë¡œ ì„¤ì • ======
input_dir = '/content/drive/MyDrive/cropped_signs_test'  # âœ… test ê¸°ì¤€ ê²½ë¡œ
reference_dir = '/content/drive/MyDrive/reference_signs'
output_dir = '/content/drive/MyDrive/test_signs'  # âœ… ê²°ê³¼ ì €ì¥ ë””ë ‰í† ë¦¬
csv_output_path = '/content/drive/MyDrive/test1.csv'  # âœ… CSV íŒŒì¼ëª…

# ====== ë””ë ‰í† ë¦¬ ì¡´ì¬ í™•ì¸ ======
assert os.path.exists(input_dir), f"âŒ ì…ë ¥ ë””ë ‰í† ë¦¬ ì—†ìŒ: {input_dir}"
assert os.path.exists(reference_dir), f"âŒ ê¸°ì¤€ ë””ë ‰í† ë¦¬ ì—†ìŒ: {reference_dir}"

# ====== í´ë” ì¤€ë¹„ ======
os.makedirs(output_dir, exist_ok=True)
os.makedirs(os.path.join(output_dir, 'unknown'), exist_ok=True)

# ====== ëª¨ë¸ ì¤€ë¹„ ======
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = resnet18(weights=ResNet18_Weights.DEFAULT)
model.fc = torch.nn.Identity()
model = model.to(device).eval()

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

def get_embedding(image_path):
    try:
        image = Image.open(image_path).convert('RGB')
        image = transform(image).unsqueeze(0).to(device)
        with torch.no_grad():
            embedding = model(image).squeeze().cpu()
        return embedding
    except UnidentifiedImageError:
        return None

# ====== í´ë˜ìŠ¤ ì„ë² ë”© ìƒì„± ======
class_embeddings = {}
for class_name in os.listdir(reference_dir):
    class_path = os.path.join(reference_dir, class_name)
    if not os.path.isdir(class_path):
        continue
    emb_list = []
    for img_file in os.listdir(class_path):
        if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            emb = get_embedding(os.path.join(class_path, img_file))
            if emb is not None:
                emb_list.append(emb)
    if emb_list:
        class_embeddings[class_name] = torch.stack(emb_list).mean(dim=0)
        os.makedirs(os.path.join(output_dir, class_name), exist_ok=True)

# ====== ì´ë¯¸ì§€ ë¶„ë¥˜ ======
results = []

for file in tqdm(os.listdir(input_dir), desc="ğŸ§  test ì„¸íŠ¸ ìœ ì‚¬ë„ ë¶„ë¥˜ ì¤‘"):
    file_path = os.path.join(input_dir, file)
    if not file.lower().endswith(('.jpg', '.jpeg', '.png')):
        continue

    emb = get_embedding(file_path)
    if emb is None:
        continue
    emb = emb.unsqueeze(0)

    similarities = {
        cls: cosine_similarity(emb.numpy(), ref.unsqueeze(0).numpy())[0][0]
        for cls, ref in class_embeddings.items()
    }
    best_class = max(similarities, key=similarities.get)
    best_score = similarities[best_class]

    label = best_class if best_score >= 0.5 else 'unknown'
    shutil.copy(file_path, os.path.join(output_dir, label, file))

    results.append({
        'filename': file,
        'class': label,
        'score': round(best_score, 4)
    })

# ====== ê²°ê³¼ ì €ì¥ ======
df = pd.DataFrame(results)
df.to_csv(csv_output_path, index=False)
print(f"\nâœ… CSV ì €ì¥ ì™„ë£Œ: {csv_output_path}")
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix, precision_recall_fscore_support
import seaborn as sns
import matplotlib.pyplot as plt

# ê²½ë¡œ ì„¤ì • (âœ… test ê¸°ì¤€ìœ¼ë¡œ ë³€ê²½)
manual_label_path = '/content/drive/MyDrive/test1_final_match.xlsx'  # ìˆ˜ë™ ë¼ë²¨ë§ íŒŒì¼
prediction_path = '/content/drive/MyDrive/test1.csv'  # ëª¨ë¸ ì˜ˆì¸¡ ê²°ê³¼

# ğŸ”¹ ì—‘ì…€ (ìˆ˜ë™ ì •ë‹µ ë¼ë²¨) ë¶ˆëŸ¬ì˜¤ê¸°
df_manual = pd.read_excel(manual_label_path)[['filename', 'match']]
df_manual.rename(columns={'match': 'true_label'}, inplace=True)

# ğŸ”¹ CSV (ëª¨ë¸ ì˜ˆì¸¡ ê²°ê³¼) ë¶ˆëŸ¬ì˜¤ê¸°
df_pred = pd.read_csv(prediction_path)[['filename', 'class']]
df_pred.rename(columns={'class': 'predicted_label'}, inplace=True)

# ğŸ”¹ ë³‘í•©
df = pd.merge(df_manual, df_pred, on='filename', how='inner')

# ğŸ”¹ í˜¼ë™ í–‰ë ¬ ê³„ì‚°
labels = sorted(df['true_label'].unique())
cm = confusion_matrix(df['true_label'], df['predicted_label'], labels=labels)

# ğŸ”¹ í˜¼ë™ í–‰ë ¬ ì‹œê°í™”
plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(cm, index=labels, columns=labels), annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix (Test Set)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# ğŸ”¹ ì •ë°€ë„, ì¬í˜„ìœ¨, F1 ê³„ì‚°
precision, recall, f1, _ = precision_recall_fscore_support(df['true_label'], df['predicted_label'], labels=labels, average=None)

print("ğŸ“Š í´ë˜ìŠ¤ë³„ Precision, Recall, F1 Score (Test Set):")
for label, p, r, f in zip(labels, precision, recall, f1):
    print(f"{label:<10} | Precision: {p:.2f} | Recall: {r:.2f} | F1 Score: {f:.2f}")

# ğŸ”¹ ì „ì²´ í‰ê· 
precision_macro = precision.mean()
recall_macro = recall.mean()
f1_macro = f1.mean()

print(f"\nğŸ“ˆ ì „ì²´ í‰ê·  Precision: {precision_macro:.2f}")
print(f"ğŸ“ˆ ì „ì²´ í‰ê·  Recall: {recall_macro:.2f}")
print(f"ğŸ“ˆ ì „ì²´ í‰ê·  F1 Score: {f1_macro:.2f}")
```
